```python
sns.lineplot(x = 'price_date', y = 'price_off_peak_var', data = agg_df)
plt.xticks(rotation = 90)
plt.show()
```

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))


# Plotting data
axs[0, 0].plot(agg_df['price_date'], agg_df['price_off_peak_var'])
axs[0, 0].set_title('Off Peak Variable Price')

axs[0, 1].plot(agg_df['price_date'], agg_df['price_peak_var'])
axs[0, 1].set_title('Peak Variable Price')

axs[0, 2].plot(agg_df['price_date'], agg_df['price_mid_peak_var'])
axs[0, 2].set_title('Mid Peak Variable Price')

axs[1, 0].plot(agg_df['price_date'], agg_df['price_off_peak_fix'])
axs[1, 0].set_title('Off Peak Fixed Price')

axs[1, 1].plot(agg_df['price_date'], agg_df['price_peak_fix'],)
axs[1, 1].set_title('Peak Fixed Price')

axs[1, 2].plot(agg_df['price_date'], agg_df['price_mid_peak_fix'])
axs[1, 2].set_title('Mid Peak Fixed Price')

# Add grid and labels to each axis for clarity
for ax in axs.flat:
    ax.grid(True)
    ax.set_xlabel('Date')
    ax.set_ylabel('Price')
    ax.tick_params(axis='x', rotation=90)
# Adjust layout to prevent overlap

plt.tight_layout()
plt.show()
```

```python
sns.countplot(data = client_df, x = 'churn')
plt.title("Class Distribution")
plt.show()
```

```python
class_counts = client_df['churn'].value_counts()

plt.pie(class_counts, labels=class_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Churn Distribution")
plt.show()
```

```python
sns.set(style="whitegrid")
ax = sns.countplot(data=client_df, x='churn')

# Add annotations on the bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',    # Get the height (count) of each bar
                (p.get_x() + p.get_width() / 2., p.get_height()),  # Position at the center of the bar
                ha='center', va='baseline', fontsize=10, color='black', xytext=(0, 5),  # Offset text slightly
                textcoords='offset points')

# Add title and labels
plt.title("Class Distribution")
plt.xlabel("Customer Churn")
plt.ylabel("Count")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 10))  # Adjust the figsize as needed

# Generate the heatmap
dataplot = sns.heatmap(
    client_df.corr(numeric_only=True), 
    annot=True, 
    fmt=".2f",  # Limit the decimal places
    annot_kws={"size": 10},  # Adjust annotation font size
    cmap="coolwarm",  # Optional: Change colormap for better readability
    cbar_kws={"shrink": 0.8}  # Optional: Shrink the color bar if needed
)

# Add title and adjust layout
plt.title("Correlation Heatmap", fontsize=16)
plt.tight_layout()  # Automatically adjust spacing to fit elements
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
add_agg_df = agg_df[['price_off_peak_var', 'price_peak_var']]
add_agg_df
```

```python
pca = PCA(n_components = 1)
x_add_agg_df = pca.fit_transform(add_agg_df)
x_add_agg_df
```

```python
type(add_agg_df)
```

```python
x_add_agg_df = pd.DataFrame(x_add_agg_df)
```

```python
agg_df = agg_df.drop(['price_off_peak_var', 'price_peak_var'],axis = 1)
```

```python
other_agg_df = pd.concat([agg_df, x_add_agg_df], axis=1, join="inner")
other_agg_df
```
